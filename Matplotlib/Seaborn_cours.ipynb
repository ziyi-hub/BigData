## types de graphiques pour une analyse bivariée

On peut voir la distribution de deux variables en même temps et donc de voir si il existe une relation entre ces deux variables.
Dans le x  et y on met le nom des colonnes, data on met le dataset à analyser et kind : quel type de graphique on souhaite.

```python
#ici on regarde le total de la facture avec les pourboires en nuage de points
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

On voit une tendance à la hausse.
On peut faire le même graph avec le paramètre hex pour obtenir des points hexagonal plus ou moins foncé selon la concentration des données.

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
#une regression
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#une regression
sns.jointplot(x='total_bill',y='tip',data=tips)
```

Ressemble à un nuage de points avec une droite de tendance en plus.

## pairplot
pairplot est une fonction qui trace les relations par paire entre toutes les colonnes numérique du dataset.
c

```python
sns.pairplot(tips)
```

On peut egalement voir la répartition en fonction de certaines variables non numériques (variables catégorielles) avec le paramètre hue. Ici on va regarder en fonction du genre des personnes en utilisant la variable 'sex'

```python
sns.pairplot(tips,hue='sex')
```

Par defaut on va avoir une fonction de densité mais on peut forcer pour avoir des histogrammes.

```python
sns.pairplot(tips,hue='sex',diag_kind="hist")
```

NB : Un histogramme est un graphique qui montre combien de fois chaque valeur apparaît dans un jeu de données à l'aide de barres. Les barres sont regroupées en bacs, qui représentent une plage de valeurs. La hauteur de chaque barre représente la fréquence des valeurs dans ce bac, ou le nombre de fois qu'une valeur apparaît dans cette plage.

D'autre part, un diagramme de densité est une courbe lisse qui montre la distribution des données. La courbe représente la proportion des données dans chaque plage, plutôt que la fréquence. Cela signifie que la hauteur de la courbe n'indique pas le nombre de fois qu'une valeur apparaît, mais plutôt la proportion des données qui se situent dans cette plage.

```python
#on peut choisir une palette de couleur
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

# Visualiser les variables categorielles
Les variables catégorielles sont des variables non numériques

## les barplots ( ou diagramme à barre)

Les barplots nous permet de voir les difference entre les hommes et les femmes sur le total de l'addition.


```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

Par defaut barplot aggrège les données sur la moyenne.
Cependant on peut lui donner un autre fonction d'aggregation pour faire son calcul.

```python
# exemple avec la variance
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

## Les countplots

Similaire au barplot sauf que la fonction d'aggrégation est explicitement que l'on compte les occurences.

```python
sns.countplot(x='sex',data=tips)
```

Il y a plus d'hommes que de femme dans le dataset.

## boxplot

La boite à moustache permet de visualiser la distribution des variables catégorielles.Cela facilite la comparaison entre les differents niveau des variables catégorielles.

On peut interpreter ce graphique comme ceci:





Les points qui sont en dehors sont des outliers ou valeurs aberrantes: c'est à dire des valeurs extrêmes  anormalement différente de la distribution d'une variable.

```python
# visualiser les bills en fonction des jours de la semaine
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
#on peut egalement ajouter une autre variable categorielle pour comparaison
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

On voit ici par exemple que les fumeurs dépensent plus que les non fumeurs le week end.

## Les stripplot

## Les matrices de Correlations

C'est un tableau contenant les coefficients de correlation entre les variables. C'est à dire si il y une relation entre les variables.
On peut les representer sous forme de graphiques.


```python
#mon tableau
tips.head()
```

```python
# on créé d'abord une matrice de correlation de toutes les données(quantitatives)
tips.corr()
```

## Les graphiques de régression

On peut afficher des modèles linéaires avec la méthode lmplot().
Cela va afficher la dispersion des données (nuages de points) avec des lignes qui indiquent la tendance de ces données.

```python
#on regarde la relation entre l'addition et les pourboires
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
#on peut comparer avec une variable catégorielle avec les couleurs
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='time',palette='coolwarm')
```

```python
#changer les marqueurs en utilisant les fonction de matplotlib
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',
           markers=['o','v'])
```

https://matplotlib.org/stable/api/markers_api.html

N'oubliez pas Seaborn etant une surcouche de Matplotlib on peut utiliser les fonctions de Matplotlib pour customiser ses graphs seaborn.

Si l'on souhaite séparer les variables catégorielles par colonnes pour avoir des graphs séparés:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
#on peut ajouter une autre variable catégorielle en ligne

sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

Egalement on peut utliser les couleurs pour encore affiner les choses. Par exemple si on veut voir la consommation les jours de la semaine et aussi en fonction du genre

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```

### ajuster les tailles et l'aspect

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=1,height=3)
```

# Couleurs et Style

## Styles

On peut avec Seaborn controler les aspects esthétiques des figures.

```python
#exemple avec un countplot
sns.countplot(x='sex',data=tips)
```

```python
#changer le background
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
# ajouter des tirets sur les axes
sns.set_style('ticks')
sns.countplot(x='sex',data=tips,palette='deep')
```

```python
#enlever le cadre
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
sns.countplot(x='sex',data=tips)
sns.despine(left=True)
```

### Taille

Souvent on utilise Matplotlib en plus de Seaborn pour controler la taille des graphiques

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

On peut utiliser set_context()pour changer les paramètres par défault :

```python
sns.set_context('poster')
sns.countplot(x='sex',data=tips,palette='coolwarm')
```
