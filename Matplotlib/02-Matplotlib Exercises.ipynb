## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
#on crée d'abord un objet figure (un espace vide pour la figure)
fig = plt.figure()
# on ajoute des axes

axes = fig.add_axes([0, 0, 1, 1])#pour ajouter des axes on passe une liste avec dans l'ordre :gauche,bas ,largeur,hauteur
#on dessine la courbe 
axes.plot(x,y)
#on met des noms aux axes
axes.set_xlabel('x') # set_ pour la methode
axes.set_ylabel('y')
axes.set_title('title')
```

## Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
# je reprends donc ma figure 
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1]) # axes principaux
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2]) # axes interieur position par rapport à l'axe princ

# La grande Figure Axes 1
ax1.plot(x, y, 'r')#en rouge
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('')

# la Figure interieure Axes 2
ax2.plot(x, y, 'r')#en rouge
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('');
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
# je reprends donc ma figure 
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # axes principaux
axes2 = fig.add_axes([0.2, 0.5, .4, .4]) # axes interieur position par rapport à l'axe princ#

# La grande Figure Axes 1
axes1.plot(x, z,'b')#en bleu
axes1.set_xlabel('X')
axes1.set_ylabel('Z')
axes1.set_xlim([30, 50])

# la Figure interieure Axes 2
axes2.plot(x, y, 'b')#en bleu
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('zoom');
axes2.set_xlim([20, 22])

```

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
# si je veux plusieurs plot je spécifie lenombre de lignes et de colonnes
fig, axes = plt.subplots(nrows=1, ncols=2)
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
# on peut également utiliser l'index
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, 'b--', linewidth=3.0)
axes[1].plot(x, z, 'r', linewidth=3.0)
fig
#ajuste les paramètres pour avoir une figure redimentionnée
plt.tight_layout()
```

```python
# on peut également utiliser l'index
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 2))

axes[0].plot(x, y, 'b', linewidth=5.0)
axes[1].plot(x, z, 'r--', linewidth=3.0)
fig
#ajuste les paramètres pour avoir une figure redimentionnée
plt.tight_layout()
```
