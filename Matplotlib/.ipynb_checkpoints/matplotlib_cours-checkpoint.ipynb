## Exemple Basique

deux méthodes pour afficher un graphique :

la méthode fonctionnelle et la méthode objet

### La méthode fonctionnelle

```python
# méthode fonctionnelle
plt.plot(x,y,'r')#on peut choisir une couleur r = red
#on peut ajouter des noms aux axes
plt.xlabel('X Axe Titre')
plt.ylabel('Y Axe Titre')
#ajouter un titre 
plt.title('Titre du graph')
```

## créer plusieurs graphiques sur la même figure

```python
# plt.subplot(nbrelignes, nbrecols, num_figure)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') #rouge en pointillé
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')#vert avec des points
```

## la méthode orientée objet

Dans la méthode orientée objet nous instantions des objets figure et ensuite appelons des méthodes sur ces objets.
Cela offre d'avantage de flexibilité et de personalisation mais elle est un peu plus difficile d'utilisation.

```python
#on crée d'abord un objet figure (un espace vide pour la figure)
fig = plt.figure()
# on ajoute des axes

axes = fig.add_axes([0.1, 0.1,0.9,0.9])#pour ajouter des axes on passe une liste avec dans l'ordre :gauche,bas ,largeur,hauteur
#on dessine la courbe 
axes.plot(x,y)
#on met des noms aux axes
axes.set_xlabel('Set X Label') # set_ pour la methode
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

## placer plusieurs figures dans un même espace

```python
#je crée ma figure vide 
fig = plt.figure()

# defini les axes 
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # axes principaux
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # axes interieur position par rapport à l'axe princ
```

```python
# je reprends donc ma figure 
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # axes principaux
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # axes interieur position par rapport à l'axe princ#

# La grande Figure Axes 1
axes1.plot(x, y, 'b')#en bleu
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('grande figure')

# la Figure interieure Axes 2
axes2.plot(y, x, 'r')#en rouge
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('petite figure');
```

## subplots()

```python
# on va utiliser une méthode similaire que le plt.figure() sauf que la cela va nous renvoyer à la fois la figure  et les axes 
fig, axes = plt.subplots()
# si je ne spécifie rien je fais ma figure comme précédemment
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('titre')
```

```python
# si je veux plusieurs plot je spécifie lenombre de lignes et de colonnes
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
#l'objet axes est un array de plusieurs axes
axes
```

```python
#on peut donc iterer dans cet array
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
# afficher la figure 
fig
```

```python
# on peut également utiliser l'index
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('figure1')
axes[1].plot(y, x)
axes[1].set_title('figure2')

fig
#ajuste les paramètres pour avoir une figure redimentionnée
plt.tight_layout()
```

## taille des figures

on peut spécifier avec Matplotlib la tailles des figure et le DPI 

* taille des figures avec l'argument figsize : tuple avec la largeur et la hauteur de la figure (en pouces)


```python
fig = plt.figure(figsize=(5,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# pareil pour les subplots
fig, axes = plt.subplots(figsize=(12,3))


axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig, axes = plt.subplots(nrows= 2, ncols = 1, figsize=(11,3))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

### Legende
On peut utilise l'argument label dans la méthode plot et ensuite utiliser la methode legend sans argument qui va me mettre automatiquement les labels en légénde.

```python
#je crée l'espace pour ma figure
fig = plt.figure()

# je rajoute les axes
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x au carre")
ax.plot(x, x**3, label="x au cube")
ax.legend()
```

il est possible de choisir l'emplacement de la legende

```python
#je crée l'espace pour ma figure
fig = plt.figure()

# je rajoute les axes
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x au carre")
ax.plot(x, x**3, label="x au cube")
ax.legend(loc=4)
#je vous laisse explorer plus cette fonction en faisant des recherches si besoin
```

### Apparence des figures

Matplotlib a beaucoup d'options pour customiser les couleurs , l'épaisseurs des lignes et les types de lignes.

**La syntaxe basique*



```python
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # ligne bleue avec des points
ax.plot(x, x**3, 'g--') # ligne verte en pointillée
```

**Avec le paramètre color**


On peut définir les couleurs par leur nom ou leur code RGB.
Le paramètre alpha indique l'opacité.

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code
```

```python
#plusieurs exemples

fig, ax = plt.subplots(figsize=(12,6))

# epaisseur de lignes
ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# type de lignes possibles ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# tiret customisés
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# marker  qui marque chaque points = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1',markersize=20)

# marker taille et couleur
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8,
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
```

### Ne montrer qu'une partie de la figure
On utilise les methodes set_ylim et set_xlim en mettant une liste en argument avec ce que le min et maximum que l'on veut voir

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("echelle par défaut")

axes[1].plot(x, x**2, x, x**3)
axes[1].set_ylim([0, 60])
axes[1].set_xlim([2, 5])
axes[1].set_title("echelle customisé")
```
